# Assignment-5(Plotting)

Refer to the tips.csv dataset. The dataset contains total 244 tips(data about customers served by a restaurant) across 7 different columns.
- totbill - Total bill (cost of the meal), including tax, in US dollars
- tip - Tip (gratuity) in US dollars
- sex - Sex of person paying for the meal (0=male, 1=female)
- smoker - Smoker in party? (0=No, 1=Yes)
- day - Sun, Sat, Thu, Fri
- time - Night and Day
- size - Size of the party

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load built-in tips dataset
df = sns.load_dataset("tips")
```

### 1. Plot a day wise boxplot of totbill column.

```python
# Create boxplot
plt.figure()
sns.boxplot(x='day', y='total_bill', data=df)

# Labels and title (as shown in image)
plt.xlabel("day")
plt.ylabel("totbill")
plt.title("Day wise Boxplot of totbill")

plt.show()
```

### 2. Plot a pariplot by setting hue parameter t.

```python
sns.pairplot(df, hue='time')
plt.show()
```

### 3. Draw a scatter plot of tip and totalbill and also allot the points according to the sex.

```python
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=df)
plt.show()
```

### 4. Plot a hitsogram of totbill for 50 bins.

```python
plt.figure()
plt.hist(df['total_bill'], bins=50)
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.title("Histogram of Total Bill")
plt.show()
```

### 5. Plot a heatmap to represent entire dataset's correlation.

```python
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap of Tips Dataset")
plt.show()
```
