```python
import pandas as pd
```

### 1. Create a DataFrame:
- Load the above dataset into a Pandas DataFrame.

```python
data = {
    "Order ID": [1001,1002,1003,1004,1005,1006,1007,1008],
    "Customer": ["Alice","Bob","Charlie","Alice","David","Charlie","Bob","Alice"],
    "Product": ["Laptop","Phone","Chair","Table","Laptop","Phone","Chair","Phone"],
    "Category": ["Electronics","Electronics","Furniture","Furniture","Electronics","Electronics","Furniture","Electronics"],
    "Quantity": [2,1,4,1,1,2,2,1],
    "Price": [800,500,100,250,800,500,100,500],
    "Total Sales": [1600,500,400,250,800,1000,200,500]
}
df = pd.DataFrame(data)
df
```

### 2. Add a New Column:
- Add a column Discount that gives a 10% discount on total sales.

```python
df["Discount"] = df["Total Sales"] * 0.10
df
```

### 3. Add a New Row:
- Add a new sales record: Order ID = 1009, Customer = "Eve", Product = "Tablet", Category = "Electronics", Quantity = 1, Price = 300, Total Sales = 300.

```python
new_row = {
    "Order ID": 1009,
    "Customer": "Eve",
    "Product": "Tablet",
    "Category": "Electronics",
    "Quantity": 1,
    "Price": 300,
    "Total Sales": 300,
    "Discount": 30
}
df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
df
```

### 4. Filter Data:
- Retrieve only the sales records where the Category is "Electronics".

```python
electronics_df = df[df["Category"] == "Electronics"]
electronics_df
```

### 5. Selection:
- Select the Customer and Total Sales columns.

```python
result = electronics_df[["Customer", "Total Sales"]]
result
```

### 6. Grouping:
- Group the data by Customer and calculate the total sales per customer.

```python
customer_sales = df.groupby("Customer")["Total Sales"].sum().reset_index()
customer_sales
```

### 7. Sorting:
- Sort the DataFrame based on Total Sales in descending order.

```python
df_sorted = df.sort_values(by="Total Sales", ascending=False)
df_sorted
```

### 8. Apply Function:
- Apply a lambda function to create a new column Sales Category, where values are labeled as "High" if Total Sales > 1000, otherwise "Low".

```python
df_sorted["Sales Category"] = df_sorted["Total Sales"].apply(
    lambda x: "High" if x > 1000 else "Low"
)
```

### 9. Save DataFrame:
- Save the final DataFrame as a CSV file named sales_data.csv.

```python
df_sorted.to_csv("sales_data.csv", index=False)
```
