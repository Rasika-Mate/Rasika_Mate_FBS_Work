# Assignment-6 (Plotting Matplotlib)

Dataset: The Cars93 dataset contains information about different car models, including their manufacturer, price, horsepower, MPG, and other attributes. You will use this dataset for the following tasks.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
cars = pd.read_csv("cars93.csv")
```

### 1. Bar Chart:
- Create a bar chart to compare the average price of cars by manufacturer.
- Label the axes and add a suitable title.
- Use different colors for different manufacturers.

```python
avg_price = cars.groupby("Manufacturer")["Price"].mean()
plt.bar(avg_price.index, avg_price.values)
plt.xlabel("Manufacturer")
plt.ylabel("Average Price")
plt.title("Average Car Price by Manufacturer")
plt.xticks(rotation=90)
plt.show()
```

### 2. Line Chart:
- Create a line chart showing the trend of horsepower for different types of cars.
- Include proper legends, labels, and a title

```python
for t in cars["Type"].unique():
    data = cars[cars["Type"] == t]
    plt.plot(data["Horsepower"].values, label=t)

plt.xlabel("Cars")
plt.ylabel("Horsepower")
plt.title("Horsepower Trend by Car Type")
plt.legend()
plt.show()
```

### 3. Pie Chart:
- Create a pie chart showing the proportion of different car types in the dataset.
- Include percentages in the chart.

```python
type_count = cars["Type"].value_counts()

plt.pie(type_count, labels=type_count.index, autopct="%1.1f%%")
plt.title("Car Type Distribution")
plt.show()
```

### 4. Scatter Plot:
- Create a scatter plot to show the relationship between horsepower and MPG(Miles Per Gallon).
- Label the axes and add a title.

```python
plt.scatter(cars["Horsepower"], cars["MPG.city"])
plt.xlabel("Horsepower")
plt.ylabel("MPG (City)")
plt.title("Horsepower vs MPG")
plt.show()
```
